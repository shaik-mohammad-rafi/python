```python
import io
import csv

sample_csv = """name,age,city
Alice,30,New York
Bob,25,Los Angeles
Charlie,35,Chicago
"""

file_like = io.StringIO(sample_csv)
reader = csv.DictReader(file_like)

rows = list(reader)
print("Sample dataset loaded from built-in csv/io:")
for row in rows:
    print(row)
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris(as_frame=True)
iris_df = iris.frame

print("Iris dataset loaded from sklearn:")
print(iris_df.head())
print("\nFeature names:", iris.feature_names)
print("Target names:", iris.target_names)
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

iris_df = iris_df.copy()
iris_df["species"] = iris_df["target"].map({i: name for i, name in enumerate(iris.target_names)})

# Histogram for each feature
iris_df[iris.feature_names].hist(bins=12, figsize=(10, 8), layout=(2, 2), color="#4C72B0", edgecolor="black")
plt.suptitle("Iris Feature Distributions")
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Scatter matrix
plt.figure(figsize=(10, 10))
scatter_matrix(iris_df[iris.feature_names], alpha=0.8, diagonal="hist", figsize=(10, 10), color="#4C72B0")
plt.suptitle("Iris Feature Scatter Matrix", y=1.05)
plt.show()

# Boxplots by species
plt.figure(figsize=(10, 6))
iris_df.boxplot(column=iris.feature_names, by="species", grid=False, rot=45)
plt.suptitle("Feature Boxplots by Species")
plt.suptitle("")
plt.xlabel("Species")
plt.ylabel("Value")
plt.show()
```
